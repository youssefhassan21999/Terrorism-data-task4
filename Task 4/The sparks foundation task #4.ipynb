```python
columns_to_work_with = ['iyear','country_txt','region_txt','provstate','city','attacktype1_txt','targtype1_txt','target1','natlty1_txt','gname','weaptype1_txt','nkill','nwound']
new_df = df[columns_to_work_with].copy()
new_df.head()
```

```python
years_count = new_df['iyear'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (20,10))
sns.barplot(x = new_df['iyear'].unique(), y = years_count)
plt.xticks(rotation = 50)
plt.xlabel('Year of attack',fontsize=20)
plt.ylabel('Number of attacks that year',fontsize=20)
plt.show()
```

```python
region_count = new_df['region_txt'].value_counts(dropna = False).sort_index()
plt.figure(figsize = (20,10))
sns.barplot(x = new_df['region_txt'].unique(), y = region_count)
plt.xlabel('Region',fontsize=20)
plt.ylabel('Number of attacks in this region',fontsize=20)
plt.show()
```

```python
pd.crosstab(new_df['iyear'], new_df['region_txt']).plot(kind='line',stacked=False,figsize=(20,10))
plt.title('Terrorist Activities By Region In Each Year',fontsize=25)
plt.xticks(rotation = 50)
plt.ylabel('Number of Attacks',fontsize=20)
plt.xlabel("Year",fontsize=20)
plt.show()
```

```python
plt.subplots(figsize=(20,10))
sns.barplot(new_df['country_txt'].value_counts()[:10].index,new_df['country_txt'].value_counts()[:10].values,)
plt.title('Top Countries Affected By Terrorism')
plt.xlabel('Countries')
plt.ylabel('Count')
plt.xticks(rotation = 50)
plt.show()
```

```python
plt.subplots(figsize=(20,10))
sns.barplot(new_df['city'].value_counts()[:10].index,new_df['city'].value_counts()[:10].values)
plt.title('Top Cities Affected By Terrorism')
plt.xlabel('Cities')
plt.ylabel('Count')
plt.xticks(rotation = 50)
plt.show()
```

```python
plt.subplots(figsize=(20,10))
sns.barplot(new_df['attacktype1_txt'].value_counts()[:10].index,new_df['attacktype1_txt'].value_counts()[:10].values)
plt.title('Top attack types')
plt.xlabel('attack types')
plt.ylabel('Count')
plt.xticks(rotation = 50)
plt.show()
```

```python
new_df[['attacktype1_txt','nkill']].groupby(["attacktype1_txt"],axis=0).sum().plot(kind='bar',figsize=(20,10),color=['darkslateblue'])
plt.xticks(rotation=50)
plt.title("Number of killed people/ attack type",fontsize=20)
plt.ylabel('Number of people',fontsize=15)
plt.xlabel('Attack type',fontsize=15)
plt.show()
```

```python
new_df['gname'].value_counts().to_frame().drop('Unknown').head(10).plot(kind='bar',color='green',figsize=(20,10))
plt.title('Top gang names')
plt.xlabel('gang names')
plt.ylabel('Count')
plt.xticks(rotation = 50)
plt.show()
```

```python
new_df[['gname','nkill']].groupby(['gname'],axis=0).sum().drop('Unknown').sort_values('nkill',ascending=False).head(10).plot(kind='bar',color='yellow',figsize=(20,10))
plt.title("Top 10 terrorist group attack by number of kills",fontsize=20)
plt.xlabel("terrorist group name",fontsize=15)
plt.ylabel("No of killed people",fontsize=15)
plt.show()
```

```python
new_df['natlty1_txt'].value_counts().to_frame().head(10).plot(kind='bar',color='green',figsize=(20,10))
plt.title('Top nationalities damged due terrorism')
plt.xlabel('nationality')
plt.ylabel('Count')
plt.xticks(rotation = 50)
plt.show()
```

```python
new_df[['natlty1_txt','nkill']].groupby(['natlty1_txt'],axis=0).sum().sort_values('nkill',ascending=False).head(10).plot(kind='bar',color='green',figsize=(20,10))
plt.title("Top 10 countires affected by terrosim / no. of deaths",fontsize=20)
plt.xlabel("country",fontsize=15)
plt.ylabel("No of killed people",fontsize=15)
plt.show()
```
